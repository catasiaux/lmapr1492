```python
from pymatgen import Structure
from jupyter_jsmol.pymatgen import quick_view
from pymatgen.ext.matproj import MPRester
from pymatgen.analysis.diffraction.xrd import XRDCalculator
import matplotlib.pyplot as plt
```

```python
with MPRester("CUtlIkfw4Pq2Hd7W") as m:
    
    # Structure for material id
    structure = m.get_structure_by_material_id("mp-753774")
    structure.to(filename="KTmO2.cif")
```

```python
xrd= XRDCalculator(wavelength=1.5406, symprec=0, debye_waller_factors=None)
#calcule le modèle de diffraction de la structure
k=xrd.get_pattern(structure,(0, 90))
#donne le modèle de diffraction du matériau sous la forme d'un graphe, le diffractogramme
xrd.get_plot(structure,(0, 31))
#permet d'obtenir toutes les intensités relatives, les intensités absolues, les indices hkl et leur multiplicité sous forme d'un dictionnaire
int_abs = xrd.get_pattern(structure,scaled = False).intensity
int_rel = xrd.get_pattern(structure,scaled = True).intensity
hkls = xrd.get_pattern(structure,scaled = False).hkls
x = xrd.get_pattern(structure,scaled = False).x
# Conclusion
print("Mon matériau est le KTmO2\n")
print("Le premier pic de diffraction se situe à un angle de ",x[0],"°, il a une multiplicité de ", hkls[0][0]['multiplicity'],", une intensité relative de ",int_rel[0]," et une intensité absolue de ",int_abs[0])
print("Les indices hkl associés sont : ",hkls[0][0]['hkl'],"\n")
print("Le deuxième pic de diffraction se situe à un angle de ",x[1],"°, il a une multiplicité de ", hkls[1][0]['multiplicity'],", une intensité relative de ",int_rel[1]," et une intensité absolue de ",int_abs[1])
print("Les indices hkl associés sont : ",hkls[1][0]['hkl'],"\n")
print("Le troisième pic de diffraction se situe à un angle de ",x[2],"°, il a une multiplicité de ", hkls[2][0]['multiplicity'],", une intensité relative de ",int_rel[2]," et une intensité absolue de ",int_abs[2])
print("Les indices hkl associés sont : ",hkls[2][0]['hkl'],"\n")
print("Pour définir l'intensité relative on fixe l'intensité du pic maximal à 100. Pour ce matériau l'intensité maximale est de : ",int_abs[0],"\n")

print('Le diffractogramme composé des trois premiers pics:')
plt.show()
```
